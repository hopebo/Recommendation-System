```python
df_genome = pd.read_csv('./data/ml-20m/genome-scores.csv', ',')
```

```python
df_genome.head(10)
```

```python
test = df_genome.set_index(['movieId', 'tagId'])
```

```python
test.head(10)
```

```python
test = test.unstack(level=-1)
```

```python
test
```

```python
mid = []
ori_mid = []
exclude = []
mid_map = {}
with open('./data/ml-1m/mid_map', 'r') as f:
    for line in f:
        line = line.rstrip().split('\t')
        try:
            corr_mid = int(line[-1])
            mid_map[int(line[0])] = corr_mid
        except:
            exclude.append(int(line[0]))
```

```python
with open('./data/ml-1m/movies.dat', 'rb') as f:
    for line in f:
        line = line.decode('latin1').split("::")
        cur_mid = int(line[0])
        if cur_mid not in exclude:
            ori_mid.append(cur_mid)
            if cur_mid in mid_map:
                mid.append(mid_map[cur_mid])
            else:
                mid.append(cur_mid)
```

```python
df_train = test.loc[mid]
```

```python
df_train['ori_mid'] = ori_mid
```

```python
df_train
```

```python
df_train.dropna(axis=0, how='any', inplace=True)
```

```python
df_train.index = df_train['ori_mid']
```

```python
df_train.drop('ori_mid', axis=1, inplace=True)
```

```python
df_train.isnull().sum()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=6).fit(df_train)
```

```python
kmeans.labels_
```

```python
res = pd.Series(kmeans.labels_, index=df_train.index)
```
